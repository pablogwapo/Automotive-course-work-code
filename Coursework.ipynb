# Question 1A

Plot engine speed(x-axis) vs torque (y-axis) using averages values for throttle at 20% and 100%

```python
speed_20 = np.array([1961,2440,3404, 3938, 4395])
torque_20 = np.array([75.5,66.5,40.8,30.4,19.89])

# combine the two arrays into a single dataset
data_20 = pd.DataFrame({'Speed_20': speed_20, 'Torque_20': torque_20})
```

```python
data_20.head()
```

```python
speed_100 = np.array([1954,2438,3403, 3931, 4601])
torque_100 = np.array([144,148,158.7,160,134])

# combine the two arrays into a single dataset
data_100 = pd.DataFrame({'Speed_100': speed_100, 'Torque_100': torque_100})
```

```python
data_100.head()
```

```python
fig,ax = plt.subplots()
ax.plot(data_20['Speed_20'], data_20['Torque_20'], marker='.', label='20% load')
ax.plot(data_100['Speed_100'], data_100['Torque_100'], marker='.', label='100% load')

# add a black dotted vertical line at x = 4000
ax.axvline(x=3931, color='black', linestyle='--')

# set the title and axis labels
ax.set_title('Engine Speed vs. Torque')
ax.set_xlabel('Engine Speed (rpm)')
ax.set_ylabel('Torque (Nm)')

# add a legend
ax.legend()

plt.show()
```

```python
# Plot engine speed(x-axis) vs torque (y-axis)
fig,ax = plt.subplots()
ax.plot(data_100['Speed_100'], data_100['Torque_100'], marker='.')

# add a black dotted vertical line at x = 4000
ax.axvline(x=3931, color='black', linestyle='--')
# set the title and axis labels
ax.set_title('Engine Speed vs. Torque')
ax.set_xlabel('Engine Speed (rpm)')
ax.set_ylabel('Torque (Nm)')

plt.show()
```

# Question 1B
Plot speed (x- axis) for power (y-axis) for 20% throttle and 100% throttle.

```python
speed_20 = np.array([1961,2440,3404, 3938, 4395])
power_20 = np.array([15.5,17,14.5,12.5,8.19])

# combine the two arrays into a single dataset
data_20 = pd.DataFrame({'Speed_20': speed_20, 'Power_20': power_20})
```

```python
speed_100 = np.array([1954,2438,3403, 3931, 4601])
power_100 = np.array([29.45, 37.9, 56.6, 64.4, 68.57])

# combine the two arrays into a single dataset
data_100 = pd.DataFrame({'Speed_100': speed_100, 'Power_100': power_100})
```

```python
# Plot engine speed(x-axis) vs torque (y-axis)
fig,ax = plt.subplots()
ax.plot(data_20['Speed_20'], data_20['Power_20'], marker='.', label = '20% load')
ax.plot(data_100['Speed_100'], data_100['Power_100'], marker='.', label = '100% load')

# set the title and axis labels
ax.set_title('Engine Speed vs Power')
ax.set_xlabel('Engine Speed (rpm)')
ax.set_ylabel('Power (kW)')

# add a legend
ax.legend()

plt.show()
```

# Question 1c

speed (x-axis)  vs engine fuel consumption(kg/s) `and` speed (x-axis) vs brake specific fuel consumption (kg / Kw) at 100% throttle 

```python
fuel_consumption = np.array([27.15, 33, 47.54, 48.8, 52.3])/1000
power_100 = np.array([29.45, 37.9, 56.6, 64.4, 68.57])
speed_100 = np.array([1954,2438,3403, 3931, 4601])

# combine the two arrays into a single dataset
data_1c = pd.DataFrame({'Speed_100': speed_100, 'Power_100': power_100, 'fuel_consumption':fuel_consumption})
data_1c['bsfc'] = data_1c['fuel_consumption'] / data_1c['Power_100']
```

```python
# Plot engine speed(x-axis) vs torque (y-axis)
fig,ax = plt.subplots()
ax.plot(data_1c['Speed_100'], data_1c['fuel_consumption'], marker='.', label = 'fuel consumption')
ax.plot(data_100['Speed_100'], data_1c['bsfc'], marker='.', label =  'bsfc')

# set the title and axis labels
ax.set_title('Fuel consumption and bsfc vs engine speed')
ax.set_xlabel('Engine Speed (rpm)')

# add a legend
ax.legend()

plt.show()
```

```python
# Plot engine speed(x-axis) vs torque (y-axis)
fig,ax = plt.subplots()
ax.plot(data_100['Speed_100'], data_1c['bsfc'], marker='.')

# set the title and axis labels
ax.set_title('bsfc vs engine speed')
ax.set_xlabel('Engine Speed (rpm)')
ax.set_ylabel('bsfc(kg/kW s)')

plt.show()
```
